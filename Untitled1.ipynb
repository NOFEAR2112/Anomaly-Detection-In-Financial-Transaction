```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
import seaborn as sns

df=pd.read_csv('Final Transactions.csv')
df
```

```python
df.head()
```

```python
df['TX_DATETIME'] = pd.DatetimeIndex(df["TX_DATETIME"]).year
df
```

```python
df["TX_DATETIME"].dtype
```

```python
df.values
```

```python
x = df.drop(["TX_FRAUD","TX_FRAUD_SCENARIO"],axis=1)
x
```

```python
y =df["TX_FRAUD"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#logistic regression model
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#accuracy
xtp=model.predict(x_train)
tda=accuracy_score(xtp,y_train)

print("accuracy:",tda)
```

```python
xtp2 = model.predict(x_test)
tda2 = accuracy_score(xtp2,y_test)

print("accuracy:",tda2)
```

```python
#testing data
print("confusion matrix for training data:")
confusion_matrix(y_test,xtp2)
```

```python
import matplotlib.pyplot as plt

# Assuming model is a scikit-learn Logistic Regression model
coefficients = model.coef_.flatten()

# Get the feature names if available (replace with your actual feature names)
feature_names = ["TRANSACTION_ID", "TX_DATETIME", "CUSTOMER_ID", "TERMINAL_ID", "TX_AMOUNT", "TX_TIME_SECONDS", "TX_TIME_DAYS", "TX_FRAUD"]

# Visualize the coefficients
plt.figure(figsize=(10, 6))
plt.bar(range(len(coefficients)), coefficients)
plt.xticks(range(len(coefficients)), feature_names, rotation='vertical')
plt.title('Logistic Regression Coefficients')
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.show()
```

```python
#using isolation forest
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
isolation_forest.fit(x_train)
```

```python
anomalies_train = isolation_forest.predict(x_train)
anomalies_test = isolation_forest.predict(x_test)
```

```python
anomalies_train_binary = np.where(anomalies_train == 1, 0, 1)
anomalies_test_binary = np.where(anomalies_test == 1, 0, 1)
```

```python
accuracy_train_isolation_forest = accuracy_score(y_train, anomalies_train_binary)
accuracy_test_isolation_forest = accuracy_score(y_test, anomalies_test_binary)
```

```python
print("Accuracy on Training Data (Isolation Forest):", accuracy_train_isolation_forest)
print("Accuracy on Testing Data (Isolation Forest):", accuracy_test_isolation_forest)
```

```python
conf_matrix_train = confusion_matrix(y_train, anomalies_train_binary)

# Confusion matrix for testing data
conf_matrix_test = confusion_matrix(y_test, anomalies_test_binary)

# Display confusion matrix
disp_train = ConfusionMatrixDisplay(conf_matrix_train, display_labels=['Normal', 'Anomaly'])
disp_test = ConfusionMatrixDisplay(conf_matrix_test, display_labels=['Normal', 'Anomaly'])

# Plot confusion matrix for training data
disp_train.plot(cmap='Blues', values_format='d')
plt.title('Confusion Matrix - Training Data (Isolation Forest)')
plt.show()

# Plot confusion matrix for testing data
disp_test.plot(cmap='Blues', values_format='d')
plt.title('Confusion Matrix - Testing Data (Isolation Forest)')
plt.show()
```

```python
plt.scatter(x_train.index, anomalies_train, c=anomalies_train, cmap='viridis')
plt.title('Isolation Forest Anomaly Scores - Training Data')
plt.xlabel('Data Point Index')
plt.ylabel('Anomaly Score')
plt.colorbar(label='Anomaly Score')
plt.show()

# Plot anomaly scores for testing data
plt.scatter(x_test.index, anomalies_test, c=anomalies_test, cmap='viridis')
plt.title('Isolation Forest Anomaly Scores - Testing Data')
plt.xlabel('Data Point Index')
plt.ylabel('Anomaly Score')
plt.colorbar(label='Anomaly Score')
plt.show()
```

```python
#kmeans clustering
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
clusters_train = kmeans.fit_predict(x_train_scaled)
clusters_test = kmeans.predict(x_test_scaled)

# Map cluster labels to binary labels (you may need to adjust this based on your data)
cluster_labels_train = np.where(clusters_train == 0, 0, 1)
cluster_labels_test = np.where(clusters_test == 0, 0, 1)

# Calculate accuracy for K-Means clustering (unsupervised)
accuracy_train_kmeans = accuracy_score(y_train, cluster_labels_train)
accuracy_test_kmeans = accuracy_score(y_test, cluster_labels_test)

print("Accuracy on Training Data (K-Means):", accuracy_train_kmeans)
print("Accuracy on Testing Data (K-Means):", accuracy_test_kmeans)
```

```python
cluster_labels_train = np.where(clusters_train == 0, 0, 1)
cluster_labels_test = np.where(clusters_test == 0, 0, 1)

# Confusion matrix for training data
conf_matrix_train = confusion_matrix(y_train, cluster_labels_train)

# Confusion matrix for testing data
conf_matrix_test = confusion_matrix(y_test, cluster_labels_test)

# Display confusion matrix
disp_train = ConfusionMatrixDisplay(conf_matrix_train, display_labels=['Normal', 'Anomaly'])
disp_test = ConfusionMatrixDisplay(conf_matrix_test, display_labels=['Normal', 'Anomaly'])

# Plot confusion matrix for training data
disp_train.plot(cmap='Blues', values_format='d')
plt.title('Confusion Matrix - Training Data (K-Means)')
plt.show()

# Plot confusion matrix for testing data
disp_test.plot(cmap='Blues', values_format='d')
plt.title('Confusion Matrix - Testing Data (K-Means)')
plt.show()

# Visualize the clustering results
plt.figure(figsize=(12, 6))

# Training data
plt.subplot(1, 2, 1)
sns.scatterplot(x=x_train.iloc[:, 0], y=x_train.iloc[:, 1], hue=cluster_labels_train, palette='viridis', legend=False)
plt.title('K-Means Clustering - Training Data')

# Testing data
plt.subplot(1, 2, 2)
sns.scatterplot(x=x_test.iloc[:, 0], y=x_test.iloc[:, 1], hue=cluster_labels_test, palette='viridis', legend=False)
plt.title('K-Means Clustering - Testing Data')

plt.show()
```
